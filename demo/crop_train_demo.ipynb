```python
from torch import nn
import torch

from common import set_seed
from preprocess.process_data_sample import preprocess_bnci2014_001_sample
from naive_implement.shallow_convnet import ShallowConvNet

set_seed()
X, y = preprocess_bnci2014_001_sample(tmin=-0.5, tmax=4)
X = X[0:1,:,:]
X,X.shape
```

```python
crop_size = 500
stride = 1

crops = []
for start in range(0, X.shape[2] - crop_size + 1, stride):
    crop = X[:, :, start:start + crop_size]  # (1, 22, 500)
    crops.append(crop)

crops = torch.cat(crops, dim=0)  # (627, 22, 500)
crops,crops.shape
```

```python
model = ShallowConvNet(X.shape) # X.shape not used
model.mean_pool = nn.Identity()
model.safe_log = nn.Identity()
model.dropout = nn.Identity()
model.final_conv = nn.Identity()
model.final_squeeze = nn.Identity()
model.eval()

logits = model(crops)

logits,logits.shape
```

the recovered logits is same as the dense forward one

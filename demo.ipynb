```python
from moabb.datasets import BNCI2014_001
import mne

# 选择数据集和任务
dataset = BNCI2014_001()

subject1 = dataset.get_data(subjects=[1])

run = subject1[1]['0train']['0']

# 1. 从 annotations 中提取事件
events, event_id = mne.events_from_annotations(run)

label_names=("left_hand", "right_hand", "feet", "tongue")
# 2. 只保留需要的事件（防止杂事件混入）
event_id = {k: v for k, v in event_id.items() if k in label_names}

# 3. 构造 Epochs（标准做法）
epochs = mne.Epochs(
    run,
    events=events,
    event_id=event_id,
    tmin=0.00,
    tmax=4.00,
    baseline=None,
    preload=True,
    verbose=False,
)

# 4. 取数据





```

```python
X = epochs.get_data()  # (n_trials, n_channels, n_times)
X.shape
```

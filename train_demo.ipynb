```python
from BNCI2014_001_dataset import BNCI2014_001_Dataset
from torch.utils.data import DataLoader
from train import train
```

```python
dataset = BNCI2014_001_Dataset(train=True, mean=None, std=None)
mean, std = dataset.get_mean_std()
val_dataset = BNCI2014_001_Dataset(train=False, mean=mean, std=std)
train_loader = DataLoader(dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
import shallow_convnet
import torch

model = shallow_convnet.ShallowConvNet(dataset.get_shape())
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
train(epochs=300, model=model, train_loader=train_loader, val_loader=val_loader, optimizer=optimizer,
      criterion=criterion)
```

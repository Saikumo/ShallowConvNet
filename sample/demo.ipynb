```python
import torch

from eeg_dataset import EEGDataset
from sample.process_data_sample import preprocess_bnci2014_001_sample

X, y = preprocess_bnci2014_001_sample()
X, y
```

```python
from train_one_epoch import train_one_epoch
from torch.utils.data import DataLoader
import shallow_convnet

model = shallow_convnet.ShallowConvNet(X.shape)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

train_loader = DataLoader(EEGDataset(X, y), batch_size=64, shuffle=True)
kappas = []


for epoch in range(2):
    train_loss, train_acc, train_kappa = train_one_epoch(model, train_loader, optimizer, criterion, torch.device("cpu"))
    print(f"kappa: {train_kappa}")
    kappas.append(train_kappa)

kappas

```
